# En español

```python
url = 'https://caaap.org.pe/2017/05/09/carta-de-tarapoto-viii-foro-social-panamazonico/'
response_es = requests.get(url)
```

```python
soup_es = BeautifulSoup(response_es.content, 'html.parser')
```

```python
text_es = soup_es.get_text()
print(text_es)
```

```python
start_phrase = "La Panamazonía está"
end_phrase = "la resistencia continúa."
```

```python
start_index = text_es.find(start_phrase)
end_index = text_es.find(end_phrase) + len(end_phrase)
```

```python
if start_index != -1 and end_index != -1:
    extracted_text_es = text_es[start_index:end_index]
    print(extracted_text_es)
else:
    print("Could not find the specified text.")
```

```python
extracted_text_es=extracted_text_es.replace('\n– ', '\n')
extracted_text_es
```

```python
lines_es = extract_lines_from_text(extracted_text_es)
```

```python
file_es_path = "../data/raw/16_Carta de Tarapoto.es"
write_lines_to_file(lines_es, file_es_path)
```
