## Import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data

```python
data = {
    'Language': ['Awajún', 'Awajún', 'Spanish', 'Spanish'],
    'Corpus Status': ['Antes', 'Ahora', 'Antes', 'Ahora'],
    'N': [378951, 422046, 460863, 526449],
    'V': [60127, 62815, 27758, 22841],
    'V1': [37366, 37741, 13563, 10168],
    'V/N': [0.16, 0.1488, 0.06, 0.0434],
    'V1/N': [0.1, 0.0894, 0.03, 0.0193]
}
df = pd.DataFrame(data)
df['Corpus Status'] = pd.Categorical(df['Corpus Status'], categories=['Antes', 'Ahora'], ordered=True)
df = df.sort_values(['Language', 'Corpus Status'])
```

## Graphs

### Awajun

```python
bars = plt.bar(df.loc[df['Language'] == 'Awajún', 'Corpus Status'], df.loc[df['Language'] == 'Awajún', 'N'], color='blue')
plt.title('Tamaño del Corpus Awajún (N)', fontsize=18)
plt.ylabel('Cantidad de palabras', fontsize=14)
plt.xticks(fontsize=12)

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{int(bar.get_height())}', ha='center', va='bottom', fontsize=12)
plt.show()
```

```python
bars = plt.bar(df.loc[df['Language'] == 'Awajún', 'Corpus Status'], df.loc[df['Language'] == 'Awajún', 'V'], color='blue')
plt.title('Tamaño del vocabulario Awajún (V)', fontsize=18)
plt.ylabel('Cantidad de palabras únicas', fontsize=14)
plt.xticks(fontsize=12)
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{int(bar.get_height())}', ha='center', va='bottom', fontsize=12)
plt.show()
```

```python
bars = plt.bar(df.loc[df['Language'] == 'Awajún', 'Corpus Status'], df.loc[df['Language'] == 'Awajún', 'V1'], color='blue')
plt.title('Hapax Legomena Awajún (V1)', fontsize=18)
plt.ylabel('Conteo de Hapax', fontsize=14)
plt.xticks(fontsize=12)

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{int(bar.get_height())}', ha='center', va='bottom', fontsize=12)

plt.show()
```

```python
bars = plt.bar(df.loc[df['Language'] == 'Awajún', 'Corpus Status'], df.loc[df['Language'] == 'Awajún', 'V/N'], color='blue')
plt.title('Tasa de crecimiento del Vocabulario Awajún (V/N)', fontsize=18)
plt.ylabel('Tasa de crecimiento', fontsize=14)
plt.xticks(fontsize=12)

# Adding value annotations
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{bar.get_height():.4f}', ha='center', va='bottom', fontsize=12)
```
